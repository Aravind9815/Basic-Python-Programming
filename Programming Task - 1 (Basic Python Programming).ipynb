```python
# Python If-Else
n = int(input().strip())

if n%2 == 0 :
    if 2 <= n <= 5:
        print("Not Weird")
    if 6<=n<=20:
        print("Weird")
    if n>20:
        print("Not Weird")
else:
    print("Weird")
```

```python
# Write a function
def is_leap(year):
    leap = False
    
    # Write your logic here
    if (year % 400 == 0) or (year%100 !=0 and year%4==0) :
        leap = True
    
    return leap

year = int(input())
print(is_leap(year))
```
